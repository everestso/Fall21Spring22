# Simple Colab Example

```python
### Includes
%matplotlib inline
import matplotlib.pyplot as plt

```

```python
Weights = [5, 2] # F(X) = 5x_0 + 2x_1
F = lambda W, X: sum([w*x for w,x in zip(W,X)])

dataX1 = range(-10, 11, 1)
dataX0 = [1]*len(dataX1)

data = (list(zip(dataX0, dataX1)))
print (data)
Y = [F(Weights, X) for X in data]
print (Y)
```

```python
plt.plot(dataX1, Y)
plt.show()
```

```python
dataX2 = [x**2 for x in dataX1]
data = list(zip(dataX0, dataX1, dataX2))
print (data)

Weights = [0, 0, 3]
Y = [F(Weights, X) for X in data]
plt.plot(dataX1, Y, "g")
plt.show()
```
